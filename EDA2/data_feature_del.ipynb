```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset2/dataset_2A_interest_growth.csv',index_col=0)
```

```python
df.sort_values(by=['거래소코드', '회계년도'], inplace=True)
```

```python
# Convert '회계년도' to datetime format
df['회계년도'] = pd.to_datetime(df['회계년도'])

# Function to check if there are at least 3 consecutive years in a group
def has_three_consecutive_years(group):
    # Sort the years
    years = group['회계년도'].dt.year.sort_values()
    # Check for at least 3 consecutive years
    return any((years.shift(-2) - years) == 2)

# Group by '거래소코드' and filter out groups that don't have 3 consecutive years
filtered_data = df.groupby('거래소코드').filter(has_three_consecutive_years)

# Show the result
filtered_data
```

```python
filtered_data.to_csv('../dataset2/data2A_첫행날리기전.csv')
```

```python
filtered_data.sort_values(by=['거래소코드', '회계년도'], inplace=True)
# Group by '거래소코드' and remove the first row of each group
filtered_data_grouped = filtered_data.groupby('거래소코드').apply(lambda x: x.iloc[1:]).reset_index(drop=True)

# Show the result
filtered_data_grouped.head(), len(filtered_data_grouped)
```

```python
filtered_data_grouped.to_csv('../dataset2/data2A_첫행날림.csv')
```
