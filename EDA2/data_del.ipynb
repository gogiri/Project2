```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset2/dataset2A_1.csv',index_col=0)
```

```python
df
```

```python
# Converting '회계년도' to datetime for easier manipulation
df['회계년도'] = pd.to_datetime(df['회계년도'], format='%Y/%m')

# Grouping by '거래소코드' and checking for consecutive 3 years of data
def has_consecutive_years(group):
    years = group['회계년도'].dt.year
    return (years.diff().fillna(1) == 1).cumsum().max() >= 3

# Filtering out groups that don't have at least 3 consecutive years of data
filtered_data = df.groupby('거래소코드').filter(has_consecutive_years)

filtered_data
```

```python
filtered_data.to_csv('../dataset2/data2A_del.csv')
```
