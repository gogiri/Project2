```python
import pandas as pd

# 연속적인 이자보상배율 확인을 위한 함수
def check_consecutive(group):
    group = group.sort_values('회계년도')  # 년도별로 정렬
    # 이자보상배율이 1 미만인 경우에 1을, 그렇지 않은 경우에 0을 할당
    group['이자보상배율_1미만'] = group['[제조]이자보상배율(이자비용)'].apply(lambda x: 1 if x < 1 else 0)
    # 연속적인 1 미만 값을 찾기 위해 그룹의 변화를 계산
    group['그룹_변경'] = group['이자보상배율_1미만'].diff().ne(0).cumsum()
    # 각 그룹 내에서 연속적으로 1 미만인 경우의 수를 계산
    return group.groupby('그룹_변경')['이자보상배율_1미만'].sum().gt(1).any()

# 회사명 별로 그룹화하고 연속적으로 이자보상배율이 1 미만인 회사 확인
companies_with_consecutive = df.groupby('회사명').filter(check_consecutive)

# 결과 확인
print(companies_with_consecutive[['회사명', '회계년도', '[제조]이자보상배율(이자비용)']].head(100))
```
