```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset/dataset_2A_del.csv',index_col=0)
```

```python
# Step 2 : 회계년도를 datetime
df['회계년도'] = pd.to_datetime(df['회계년도'], format='%Y/%m')
df_sorted = df.sort_values(by=['거래소코드', '회계년도'])

# Step 3 : 감사의견코드 중 DI, DS DU, AG, QS가 존재하면 1 아니면 0
df_sorted['y'] = df_sorted['감사의견코드'].apply(lambda x: 1 if x in ['DI', 'DS', 'DU', 'AG', 'QS'] else 0)
```

```python
# Step 4: 
def remove_post_default_data(group):
    # 최초 부도 년도 설정
    default_year = group[group['y'] == 1]['회계년도'].min()
    
    # default 값이 nan이 아니면 작거나 같은 데이터만 가져옴
    if not pd.isnull(default_year):
        group = group[group['회계년도'] <= default_year]
    
    return group

# 회사명으로 그룹바이 후 함수 적용
df_filtered = df_sorted.groupby('거래소코드').apply(remove_post_default_data).reset_index(drop=True)

df_filtered.head()
```

```python
df_filtered.to_csv('../dataset/dataset_2A_label_1.csv')
```
