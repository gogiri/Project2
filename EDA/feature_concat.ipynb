```python
import pandas as pd
```

```python
df1 = pd.read_csv('../dataset/dataset_2A_label.csv', index_col= 0)
df2 = pd.read_csv('../dataset/dataset_2A_plus_feature.csv', index_col= 0)
```

```python
df1['회계년도']=pd.to_datetime(df1['회계년도'],format='%Y-%m-%d')
df2['회계년도']=pd.to_datetime(df2['회계년도'],format='%Y/%m')
```

```python
df3=df2[['회사명','거래소코드','회계년도','자본잠식률','이자보상배율']]
```
