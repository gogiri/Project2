```python
import pandas as pd
```

```python
df1 = pd.read_csv('../dataset_kosdaq/konex_ratio_total_v1.csv', encoding='utf-8')
df2 = pd.read_csv('../dataset_kosdaq/ksq_total_ver1.csv')
```

```python
df1.rename(columns={'회사명_x':'회사명'},inplace=True)
df1.rename(columns={'영업이익증가율_x':'영업이익증가율'},inplace=True)
df1.rename(columns={'총자본증가율_x':'총자본증가율'},inplace=True)
df1.rename(columns={'순이익증가율_x':'순이익증가율'},inplace=True)
df1.rename(columns={'자기자본증가율_x':'자기자본증가율'},inplace=True)
df2.rename(columns={'이자보상배율(이자비용)':'이자보상배율'},inplace=True)
```

```python
df1.drop(columns='회사명_y',inplace=True)
df1.drop(columns='총자본증가율_y',inplace=True)
df1.drop(columns='영업이익증가율_y',inplace=True)
df1.drop(columns='순이익증가율_y',inplace=True)
df1.drop(columns='자기자본증가율_y',inplace=True)
```

```python
col1 = df1.columns
col2 = df2.columns
```

```python


# list1에는 있지만 list2에는 없는 요소 찾기
only_in_list1 = [item for item in col1 if item not in col2]

# list2에는 있지만 list1에는 없는 요소 찾기
only_in_list2 = [item for item in col2 if item not in col1]

only_in_list1, only_in_list2

```

```python
df1.drop(columns=only_in_list1, inplace=True)
df2.drop(columns=only_in_list2, inplace=True)
```

```python
import pandas as pd


df1['회계년도'] = pd.to_datetime(df1['회계년도'], format="%Y-%m-%d")

df2['회계년도']=pd.to_datetime(df2['회계년도'],format='%Y/%m')
```

```python
df2
```

```python
# Merging the two datasets on '회사명' and '회계년도'
merged_dataset = pd.concat([df1, df2])
```

```python
merged_dataset.to_csv('../dataset_kosdaq/kosdaq_konex_concat.csv')
```
